### Распределение рейтингов

## Рекомендации на основании количества рейтингов

Посчитаем количество оценок у каждой книги

```python
from pyspark.sql.functions import desc

rating_count = ratings.groupby('ISBN').count()
rating_count = rating_count.sort(desc("count"))
rating_count.limit(5).toPandas()
```

Книга с идентификатором "0971880107" имеет больше всего оценок. Посмотрим информацию о 5 наиболее оцениваемых книгах.

```python
most_rated_books = sql.createDataFrame([['0971880107'], ['0316666343'], ['0385504209'], ['0060928336'], ['0312195516']], ['ISBN'])
most_rated_books_summary = books.join(most_rated_books, on=['ISBN']).sort(desc("ISBN"))
most_rated_books_summary.toPandas()
```

Книга, получившая наибольшее количество оценок в этом наборе данных - Wild Animus Rich Shapero. Все эти книги - романы. Можно предположить, что это все популярные романы. И если кому-то нравится «Wild Animus»,то возможно ему понравится «The Lovely Bones: The Novel».

## Рекомендации основанные на корреляции

Найдем средний рейтинг и общее количетсво оценок для каждой книги

```python
import pyspark.sql.functions as func
average_rating = ratings.groupBy("ISBN").agg(func.mean('Book-Rating').alias("bookRating"), 
                                             func.count('ISBN').alias("ratingCount"))
average_rating = average_rating.sort(desc("ratingCount"))
average_rating.limit(5).toPandas()
```

#### Факт: 

Наиболее оцениваемая книга в этой наборе имеет очень низкую оценку. Поэтому основываясь только на количетсве оценок, мы доспутили бы здесь ошибку.

#### Уберем из данных книги, у которых меньше 200 оценок, и пользователей, которые оценили меньше 100 книг.

```python
from pyspark.sql.functions import col

counts1 = ratings.groupby('User-ID').count()
good_user =  counts1.where(col('count') >= 200)
ratings = ratings.where(col('User-ID').isin(good_user.select("User-ID").rdd.flatMap(lambda x: x).collect()))

counts2 = ratings.groupby('ISBN').count()
good_book =  counts2.where(col('count') >= 100)
ratings = ratings.where(col('ISBN').isin(good_book.select("ISBN").rdd.flatMap(lambda x: x).collect()))

print(ratings.count(), len(ratings.columns))
```

Осталось 13793 записей

### Матрица рейтингов

Преобразуем имеющую таблицу с рейтингами в матрицу. Она будет разреженной, т.к. не все пользователи оценили все книги

Посмторим на содержимое этой таблицы

```python
ratings.limit(20).toPandas()
```

Посмотроим эту матрицу. По строкам будут распологаться пользователи, а по столбцам - книги.

```python
ratings_pivot = ratings.groupby('User-ID').pivot('ISBN').min('Book-Rating')
print(ratings_pivot.count(), len(ratings_pivot.columns))

ratings_pivot.limit(20).toPandas()
```

Найдем книги, книги которые хорошо коррелируют со второй по популярности книгой - «The Lovely Bones: The Novel». Это история девочки-подростка, которая после своей смерти наблюдает за ее семьей и друзьми, которые изо всех сил стараются продолжать свою жизнь.

```python
from pyspark.ml.stat import Correlation


bones_book = '0316666343'
corr_values = {}
for book in ratings_pivot.columns[1:]:
    corr_values[book] = ratings_pivot.stat.corr(bones_book, book)
```

```python
import operator
sorted_corr_values = sorted(corr_values.items(), key=operator.itemgetter(1))
```

Выберем 9 наиболее подходящих книг

```python
books_corr_to_bones = sc.parallelize([(k,)+(v,) for k,v in dict(sorted_corr_values[-10:-1]).items()]).toDF(['ISBN','corr'])
```

```python
books_corr_to_bones.limit(9).toPandas()
```

```python
books_corr_to_bones_info = books.join(books_corr_to_bones, on=['ISBN']).sort(desc("ISBN"))
books_corr_to_bones_info.limit(9).toPandas()
```
